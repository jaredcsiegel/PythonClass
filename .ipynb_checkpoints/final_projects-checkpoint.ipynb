### Step 0: imports

```python
# these should be familar
import scipy
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

# this package will allow us to make progress bars
from tqdm import tqdm

# We can beautify our plots by changing the matpltlib setting a little
plt.rcParams['font.size'] = 18
matplotlib.rcParams['axes.linewidth'] = 2
matplotlib.rcParams['font.family'] = "serif"
```

### Step 1: fetch the data

```python
def load_TESS_data(target, cutout_size=20):
    """
    Download TESS images for a given target.
    
    inputs:
        target: str, same of star (ex. TOI-700)
        cutout_size: int, size of the returned cutout 
                     (in units of pixels)
    outputs:
        time: numpy 1D array
        flux: numpy 3D array (time x cutout_size x cutout_size)
    """
    import lightkurve as lk
    
    # find all available data
    search_result = lk.search_tesscut(target)
    
    # select the first available observing sector
    search_result[search_result.mission == search_result.mission[0]]

    # download it
    tpf = search_result.download_all(cutout_size=cutout_size)[0]
    
    # convert it to a numpy array format
    flux = np.array([
        tpf.flux[i].value for i in range(len(tpf.flux)) if tpf.flux[i].value.min()>0
    ])
    flux_err = np.array([
        tpf.flux_err[i].value for i in range(len(tpf.flux_err)) if tpf.flux[i].value.min()>0
    ])

    time = np.array([ tpf.time.value[i] for i in range(len(tpf.flux)) if tpf.flux[i].value.min()>0 ])
    
    return time, flux, flux_err
```

```python
target = 'Gaia_DR2279382060625871360'

time, flux, flux_err = load_TESS_data(target)
```

### Step 2: visualize the data

```python
def plot_cutout(image, logscale = True, vrange = [1,99]):
    """
    Plot 2D image.
    
    inputs:
        image: 2D numpy array
        logscale: bool, if True plots np.log10(image)
        vrnge: list, sets range of colorbar 
               (based on percentiles of 'image') 
    """
    if logscale:
        image = np.log10(image)
        
    q = np.percentile(image, q = vrange)
    plt.imshow( image, vmin = q[0], vmax = q[1] )    
    plt.show()
```

```python
plot_cutout(np.mean(flux,axis=0))
```

### Step 3: point-spread function

```python
def psf(x,y,sigma,scale,bkg,xmu=0,ymu=0):
    """
    Symmetric 2D point-spread function (PSF)
    
    inputs:
        x: np.ndarray, where to evaluate the PSF
        y: np.ndarray, where to evaluate the PSF
        sigma: float, width of the PDF
        scale: float, integrated area under the PSF
        bkg: float, background flux level
        xmu: float, center of PSF in x
        ymu: float, center of PSF in y
    outputs:
        PSF evalauted at x and y
    """
    exponent = (np.power(x-xmu,2) + np.power(y-ymu,2) ) / sigma**2
    
    return scale * np.exp(-exponent) / (2*np.pi*sigma**2) + bkg
```

```python
# lets plot one
x = np.linspace(0,10,50)
y = np.linspace(0,10,50)
X, Y = np.meshgrid(x, y)
Z = psf(X,Y,sigma=1,scale=1,bkg=0,xmu=5,ymu=5)

plot_cutout(Z,logscale=False)
```

### Step 4: photometry

```python
def residual(theta, x, y, z, z_err):
    sigma, scale, bkg, xmu, ymu = theta
    
    model = psf(x,y,sigma=sigma,scale=scale,bkg=bkg,xmu=xmu,ymu=ymu)
    
    return np.sum(np.power(model - z,2)/z_err**2)
```

```python
# where to evaluate out PSF
x = np.arange(flux.shape[1])
y = np.arange(flux.shape[2])
X,Y = np.meshgrid(x,y)
```

```python
# lets evaluate the model for a set of parameters
theta = [
    1,  # sigma
    200,  # scale
    0,  # bkg
    10, # xmu
    10, # ymu
]

residual(theta, X, Y, flux[0]/np.mean(flux), flux_err[0]/np.mean(flux))
```

```python
# let's vary the 'scale' parameter to see what works best
scales = np.linspace(0,1000,10)
residuals = []
for scale in scales:
    theta = [
        1,  # sigma
        scale,  # scale
        0,  # bkg
        10, # xmu
        10, # ymu
    ]

    residuals.append( residual(theta, X, Y, flux[0]/np.mean(flux), flux_err[0]/np.mean(flux)) )
```

```python
plt.scatter(scales,residuals)
_ = plt.xlabel('Scale')
_ = plt.ylabel('$\chi^2$')
```

### Step 4: making a light curve

```python
import scipy
popts = []
for i,f in tqdm(enumerate(flux)):
    # total flux?
    out = scipy.optimize.minimize(residual, 
                              x0 = [
                                    1, 
                                    6*(f/np.mean(flux)).max(),  
                                    np.median(f/np.mean(flux)),  
                                    10,  
                                    10],
                              args = (X,Y,f/np.mean(flux),flux_err[i]/np.mean(flux)),
                              bounds = [
                                     (.5,10),
                                     ( 10 ,10000),
                                     (1e-10,np.inf),
                                     (1e-10,50),
                                     (1e-10,50),
                             ])
    popts.append( out.x )
popts = np.array(popts)
```

```python
plt.scatter(time-time.min(), popts[:,1] )
_ = plt.xlabel('Time [days]')
_ = plt.ylabel('Normalized flux')
```

### Step 5: finding the pulsation periods

```python
from astropy.timeseries import LombScargle

periods   = 10**np.linspace(np.log10(2),2,1000)
frequency = 1/periods
power = LombScargle(time,popts[:,1]).power(frequency)
```

```python
plt.plot(periods,power)
_ = plt.xlabel('Period [days]')
_ = plt.ylabel('Power')
plt.xscale('log')
```

### Step 6: repeat this process on mulitple stars

```python
def extract_lightcurve(time, flux, flux_err):
    popts = []
    for i,f in tqdm(enumerate(flux)):
        # total flux?
        out = scipy.optimize.minimize(residual, 
                                  x0 = [1, 
                                        6 * (f/np.mean(flux)).max(),  
                                        np.median(f/np.mean(flux)),  
                                        10,  
                                        10],
                                  args = (X,Y,f/np.mean(flux),flux_err[i]/np.mean(flux)),
                                  bounds = [
                                     (.5,10),
                                     ( 10 ,10000),
                                     (1e-10,np.inf),
                                     (1e-10,50),
                                     (1e-10,50),
                                 ])
        popts.append( out.x )
    popts = np.array(popts)
    
    return popts

def get_max_period(x, y, plot_it=False):
    periods = 10**np.linspace(np.log10(2),2,1000)
    frequency = 1/periods
    power = LombScargle(x,y).power(frequency)
    
    if plot_it:
        plt.plot(periods,power)
        plt.xscale('log')
        _ = plt.xlabel('Period [days]')
        _ = plt.ylabel('Power')
        plt.show()
    
    return periods[np.argmax(power)]

def plot_lightcurve(x,y):
    plt.scatter(x, y)
    _ = plt.xlabel('Time [days]')
    _ = plt.ylabel('Normalized flux')
    plt.show()
```

```python
# let's calculate the lightcurves for these stars (should take < 2 minutes)
lightcurves = []

targets = [
    'Gaia_DR2279382060625871360',
    'Gaia_DR2422923956470923008',
    'Gaia_DR2270272675510149504',
    'Gaia_DR25871738271033526016',
    'Gaia_DR25334506135119058304',
    'Gaia_DR22058374144759464064',
    'Gaia_DR2508915489570374272',
    'Gaia_DR25337764640889249536',
    'Gaia_DR25937099633141128448',
    'Gaia_DR25238808628736339584',
]

for target in targets:
    
    time, flux, flux_err = load_TESS_data(target)
    
    plot_cutout(np.mean(flux,axis=0))    

    popts = extract_lightcurve(time, flux, flux_err)
    
    plot_lightcurve(time-time.min(), popts[:,1])
    
    lightcurves.append( [time, popts[:,1]] )
```

```python
# now let's calculate the pulsation periods
periods = []
for l in lightcurves:
    time, lc = l
    periods.append( get_max_period(time, lc) )
```

### Step 7: compare to *Gaia*

```python
# load the table with pandas
gaia = pd.read_csv('data/gaia_variables.csv')
```

```python
gaia.head()
```

```python
# let's add luminosity to the Gaia table
import astropy.units as u
import astropy.constants as const

SA = 4 * np.pi * np.power(gaia.radius_val.values * u.R_sun,2)
F  = const.sigma_sb * np.power(gaia.teff_val.values * u.K,4)
L  = F * SA
L  = L.to(u.L_sun)

gaia.loc[:,'L'] = L.value
```

```python
# now let's find the Gaia pulsation period, parallax, and luminosity for our stars
gaia_periods  = []
gaia_parallax = []
gaia_L        = []

for t in targets:
    slc = gaia[gaia.source_id==int(t.split('DR2')[1])]
    gaia_periods.append(slc.pf.values[0])
    gaia_parallax.append(slc.parallax.values[0])
    gaia_L.append(slc.L.values[0])
    
gaia_periods  = np.array(gaia_periods)
gaia_parallax = np.array(gaia_parallax)
gaia_L        = np.array(gaia_L)
```

```python
# what do we compare?
plt.plot([1,35],[1,35],color='k',linestyle='--',zorder=-1)
plt.scatter(gaia_periods,periods,s=100)
_ = plt.xlabel('Gaia period [days]')
_ = plt.ylabel('TESS period [days]')
```

```python
# what about the outlier? 
# it seems Gaia's solution was an alias (2x the period) of the strongest pulsation mode
plot_lightcurve(lightcurves[-1][0],lightcurves[-1][1])
_ = get_max_period(lightcurves[-1][0],lightcurves[-1][1],True)
```
